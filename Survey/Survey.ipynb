```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('survey.csv')
```

```python
for column in df.columns:
    print(f"Field: {column}")
    print(df[column].dropna().unique())  
    print("\n" + "-"*50 + "\n")
```

```python
# Plot for each field
for column in df.columns:
    if df[column].dtype == np.int64:
        df[column].plot.hist()
        plt.title(column)
        plt.show()
    elif df[column].dtype == object:
        df[column].value_counts(ascending=True).plot.barh()
        plt.title(column)
        plt.show()
    else:
        print(f"Field {column} is not numeric or string")
        print(df[column].dropna().unique())
        print("\n" + "-"*50 + "\n")
```
